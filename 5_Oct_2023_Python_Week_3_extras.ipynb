```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

from IPython.display import Image
```

# Line Plot

```python
x = np.arange(2,21)
x
```

```python
y = x*3
y
```

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.plot(x,y,color='r' , linewidth=2,linestyle=':')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=np.arange(1,10)
x
```

```python
y1=x*2
y2=x*3
```

```python
plt.plot(x,y1,color='g',linewidth=2,linestyle=":")
plt.plot(x,y2,color='r',linewidth=1.5,linestyle='-.')
plt.xlabel('Multiple Plots')
plt.ylabel('Red and Green')
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
y1=[8,3,7,1,4,9,3,6,0]
y2=[1,2,3,8,6,5,3,9,6]
```

```python
plt.scatter(x,y1,marker='*',c='orange',s=200)
plt.scatter(x,y2,marker='.',c='red',s=100)
plt.show()
```

# Bar Plot

```python
student = {"Julia":93,"Matt":23,"Sam":67}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.figure(figsize=(15,7))
plt.bar(names,values)
plt.show()
```

```python
plt.barh(names,values,color='pink')
plt.show()
```
